#Creating catalog and schema in Databricks

```python
# Create Catalog if not exists

spark.sql(f"CREATE CATALOG IF NOT EXISTS airbnb")
```

```python
# Create schema if not exists

spark.sql(f"CREATE SCHEMA IF NOT EXISTS airbnb.bronze")
spark.sql(f"CREATE SCHEMA IF NOT EXISTS  airbnb.silver")
spark.sql(f"CREATE SCHEMA IF NOT EXISTS  airbnb.gold")
```
